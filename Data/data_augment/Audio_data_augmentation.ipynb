```python
%matplotlib inline
import matplotlib.pyplot as plt, IPython.display as ipd, numpy
import librosa, librosa.display
import os
import numpy as np
import soundfile as sf
```

```python
dir = os.listdir('C:/Users/ADmin/Desktop/samples/hihat')
```

```python
audio_list = []

for i in dir:
    
    audio = ('C:/Users/ADmin/Desktop/samples/hihat/' + i)
    audio_list.append(audio)
```

```python
#stretch Data
```

```python
for num, i in enumerate(audio_list): 
    x, sr = librosa.load(i,sr=None)
    factor = 1.4
    x_stch = librosa.effects.time_stretch(x,factor)
    
    stch_siz = x_stch.shape[0]
    siz = x.shape[0]
    
    if stch_siz > siz:
        x_stch = x_stch[stch_siz//2-siz//2:stch_siz//2+siz//2+1]
    elif stch_siz < siz:
        x_stch = np.append(np.zeros(siz//2-stch_siz//2), x_stch)
        x_stch = np.append(x_stch, np.zeros(siz//2-stch_siz//2))
    else:
        x_stch = x_stch
    
    dire = "C:/Users/ADmin/Desktop/data_augment/hihat/"+str(factor)+"_hihat_"+str(num)+".wav"
    sf.write(dire, x_stch, sr)
```
